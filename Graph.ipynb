```python
import plotly.graph_objects as go
import plotly.express as px
import pandas as pd

# load dataset

dataF1 = input('Enter the name of the first DataFile ')
dataF2 = input('Enter the name of the second DataFile ')
df = pd.read_csv(dataF1)
df1 = pd.read_csv(dataF2)
# create figure
fig = go.Figure()
fig1 = go.Figure();
# Add surface trace
fig.add_trace(go.Surface(z=df.values.tolist(), colorscale="Viridis"))
fig1.add_trace(go.Surface(z = df1.values.tolist(), colorscale = "Viridis"))


################################################################################################

# Update plot sizing
fig.update_layout(
    width=800,
    height=900,
    autosize=False,
    margin=dict(t=0, b=0, l=0, r=0),
    template="plotly_white",
    
    
    #Title MAIN
    title=go.layout.Title(
        text= dataF1 + " Abort Distribution",
        xref="paper",
        x=0.5
    ),
    
    #XAXIS
    xaxis=go.layout.XAxis(
        title=go.layout.xaxis.Title(
            text="AHDISAHIDSHI",
            font=dict(
                family="Courier New, monospace",
                size=18,
                color="#7f7f7f"
            )
        )
    ),  
)
# Update plot sizing
fig1.update_layout(
    width=800,
    height=900,
    autosize=False,
    margin=dict(t=0, b=0, l=0, r=0),
    template="plotly_white",
    
    
    #Title MAIN
    title=go.layout.Title(
        text= dataF2 + " Abort Distribution",
        xref="paper",
        x=0.5
        
    ),
    
    #XAXIS
    xaxis=go.layout.XAxis(
        title=go.layout.xaxis.Title(
            text="AHDISAHIDSHI",
            font=dict(
                family="Courier New, monospace",
                size=18,
                color="#7f7f7f"
            )
        )
    ),  
)

################################################################################################

# Update 3D scene options
fig.update_scenes(
    aspectratio=dict(x=1, y=1, z=0.7),
    aspectmode="manual"
)
fig1.update_scenes(
    aspectratio=dict(x=1, y=1, z=0.7),
    aspectmode="manual"
)

################################################################################################
# Add dropdown
fig.update_layout(
    updatemenus=[
        go.layout.Updatemenu(
            buttons=list([
                dict(
                    args=["type", "surface"],
                    label="3D Surface",
                    method="restyle"
                ),
                dict(
                    args=["type", "heatmap"],
                    label="Heatmap",
                    method="restyle"
                )
            ]),
            direction="down",
            pad={"r": 10, "t": 10},
            showactive=True,
            x=0.12,
            xanchor="left",
            y=1.105,
            yanchor="top"
        ),
        
        
        go.layout.Updatemenu(
            buttons=list([
                dict(
                    args=["colorscale", "Viridis"],
                    label="Viridis",
                    method="restyle"
                ),
                dict(
                    args=["colorscale", "Cividis"],
                    label="Cividis",
                    method="restyle"
                ),
                dict(
                    args=["colorscale", "Blues"],
                    label="Blues",
                    method="restyle"
                ),
                dict(
                    args=["colorscale", "Greens"],
                    label="Greens",
                    method="restyle"
                ),
            ]),
            direction="down",
            pad={"r": 10, "t": 10},
            showactive=True,
            x=0.12,
            xanchor="left",
            y=1.06,
            yanchor="top"
        ),
        
    ]      
)
fig1.update_layout(
    updatemenus=[
        go.layout.Updatemenu(
            buttons=list([
                dict(
                    args=["type", "surface"],
                    label="3D Surface",
                    method="restyle"
                ),
                dict(
                    args=["type", "heatmap"],
                    label="Heatmap",
                    method="restyle"
                )
            ]),
            direction="down",
            pad={"r": 10, "t": 10},
            showactive=True,
            x=0.12,
            xanchor="left",
            y=1.105,
            yanchor="top"
        ),
        
        
        go.layout.Updatemenu(
            buttons=list([
                dict(
                    args=["colorscale", "Viridis"],
                    label="Viridis",
                    method="restyle"
                ),
                dict(
                    args=["colorscale", "Cividis"],
                    label="Cividis",
                    method="restyle"
                ),
                dict(
                    args=["colorscale", "Blues"],
                    label="Blues",
                    method="restyle"
                ),
                dict(
                    args=["colorscale", "Greens"],
                    label="Greens",
                    method="restyle"
                ),
            ]),
            direction="down",
            pad={"r": 10, "t": 10},
            showactive=True,
            x=0.12,
            xanchor="left",
            y=1.06,
            yanchor="top"
        ),   
    ]
)
################################################################################################

# Add annotation
fig.update_layout(
    annotations=[
        go.layout.Annotation(text="Trace type:", showarrow=False,
                             x=0.02, y=1.085, yref="paper", align="left"),
        
        
        go.layout.Annotation(text="Colorscale:", showarrow=False,
                             x=0.02, y=1.040, yref="paper", align="left"),
    ]
)
fig1.update_layout(
    annotations=[
        go.layout.Annotation(text="Trace type:", showarrow=False,
                             x=0.02, y=1.085, yref="paper", align="left"),
        
        
        go.layout.Annotation(text="Colorscale:", showarrow=False,
                             x=0.02, y=1.040, yref="paper", align="left"),
    ]
)

################################################################################################
fig.show()
fig1.show()
```
